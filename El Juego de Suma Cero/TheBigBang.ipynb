```python
import pandas as pd
```

```python
class Objeto:
    def __init__(self, denominacion, valor ):
        self.denominacion = denominacion
        self.valor = valor
        self.objeto_opuesto = self.valor*(-1)
        self.objeto = []



        
        

        
            

    
class Sistema:
    def __init__(self ):
        self.n_de_elementos = 0
        self.lista_de_objetos = []
        self.n__de_relaciones = 0
        self.entropia = 0
        self.dimensiones = []

    def añadir_objetos(self, objeto):
        self.lista_de_objetos.append({objeto.denominacion:[objeto.valor]})
        self.lista_de_objetos.append({"-" + objeto.denominacion:[objeto.objeto_opuesto]})      
        self.n_de_elementos+=2
        self.entropia += 1
        
  

    def tiempo(self, n__de_acciones, n_de_ciclos):
        return n__de_acciones/n_de_ciclos
    
    def mostrar_elementos(self):
        for objeto in self.lista_de_objetos:
            print(objeto)

```

```python
Hagasé_la_luz = Objeto("uno", 5)





```

```python
sistema = Sistema()
sistema.añadir_objetos(Hagasé_la_luz)
```
